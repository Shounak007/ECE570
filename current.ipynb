```python



import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from generative_models.timegan.utils import (
    extract_time,
    rnn_cell,
    random_generator,
    batch_generator,
)

tf.compat.v1.disable_eager_execution()

def timegan(ori_data, parameters):
    # Here we initalize our data visualization
    tf.compat.v1.reset_default_graph()

    # Parameters
    no, seq_len, dim = np.asarray(ori_data).shape
    ori_time, max_seq_len = extract_time(ori_data)

    # Normalization funciton used to determines our accuracy
    def MinMaxScaler(data):
        min_val = np.min(np.min(data, axis=0), axis=0)
        max_val = np.max(np.max(data, axis=0), axis=0)
        return (data - min_val) / (max_val - min_val + 1e-7), min_val, max_val
    ori_data_norm, min_val, max_val = MinMaxScaler(ori_data)

    # Placeholder setup
    X = tf.compat.v1.placeholder(tf.float32, [None, max_seq_len, dim], name="myinput_x")
    Z = tf.compat.v1.placeholder(tf.float32, [None, max_seq_len, dim], name="myinput_z")
    T = tf.compat.v1.placeholder(tf.int32, [None], name="myinput_t")

    # RNN Netowrk based off of gan.py
    def embedder(X, T):
        with tf.compat.v1.variable_scope("embedder", reuse=tf.compat.v1.AUTO_REUSE):
            e_cell = tf.compat.v1.nn.rnn_cell.MultiRNNCell(
                [rnn_cell(parameters["module"], parameters["hidden_dim"]) for _ in range(parameters["num_layer"])]
            )
            e_outputs, _ = tf.compat.v1.nn.dynamic_rnn(e_cell, X, dtype=tf.float32, sequence_length=T)
            return tf.compat.v1.layers.dense(e_outputs, parameters["hidden_dim"], activation=tf.nn.sigmoid)

    def recovery(H, T):
        with tf.compat.v1.variable_scope("recovery", reuse=tf.compat.v1.AUTO_REUSE):
            r_cell = tf.compat.v1.nn.rnn_cell.MultiRNNCell(
                [rnn_cell(parameters["module"], parameters["hidden_dim"]) for _ in range(parameters["num_layer"])]
            )
            r_outputs, _ = tf.compat.v1.nn.dynamic_rnn(r_cell, H, dtype=tf.float32, sequence_length=T)
            return tf.compat.v1.layers.dense(r_outputs, dim, activation=None)

    # Training loop setup
    H = embedder(X, T)
    X_tilde = recovery(H, T)
    E_loss_T0 = tf.losses.mean_squared_error(X, X_tilde)
    E0_solver = tf.compat.v1.train.AdamOptimizer().minimize(E_loss_T0)

    # Run session
    sess = tf.compat.v1.Session()
    sess.run(tf.compat.v1.global_variables_initializer())

    for itt in range(parameters["iterations"]):
        X_mb, T_mb = batch_generator(ori_data_norm, ori_time, parameters["batch_size"])
        _, step_e_loss = sess.run([E0_solver, E_loss_T0], feed_dict={X: X_mb, T: T_mb})
        if itt % parameters["print_every_n_iters"] == 0:
            print("Iteration:", itt, "Embedding loss:", step_e_loss)

    # Generate synthetic data
    H_gen = sess.run(H, feed_dict={X: ori_data_norm, T: ori_time})
    generated_data_norm = sess.run(X_tilde, feed_dict={H: H_gen, T: ori_time})
    generated_data_norm = [generated_data_norm[i, : ori_time[i], :] for i in range(no)]
    generated_data_norm = np.array(generated_data_norm)

    # Denormalize the data
    generated_data = generated_data_norm * (max_val - min_val + 1e-7) + min_val

    sess.close()
    return generated_data

# Sample data and parameters
no, seq_len, dim = 100, 24, 5  # Number of sequences, length, and features
ori_data = np.random.rand(no, seq_len, dim)  # Create synthetic original data
parameters = {
    "module": "gru",
    "hidden_dim": 10,
    "num_layer": 3,
    "iterations": 100,
    "batch_size": 16,
    "print_every_n_iters": 10
}


generated_data = timegan(ori_data, parameters)

# Additional Data Visualization

# 1. Plot comparison of multiple sequences and features
num_sequences_to_plot = 5
features_to_plot = [0, 1, 2]  # Plot first three features

for feature_idx in features_to_plot:
    plt.figure(figsize=(12, 6))
    for sample_idx in range(num_sequences_to_plot):
        plt.plot(ori_data[sample_idx, :, feature_idx], label=f"Original Seq {sample_idx}", alpha=0.6)
        plt.plot(generated_data[sample_idx][:, feature_idx], label=f"Synthetic Seq {sample_idx}", linestyle="dashed", alpha=0.6)
    plt.title(f"Comparison of Original and Synthetic Data (Feature {feature_idx})")
    plt.xlabel("Time Step")
    plt.ylabel("Value")
    plt.legend()
    plt.show()

# 2. Histograms of feature distributions
for feature_idx in range(dim):
    plt.figure(figsize=(8, 4))
    sns.kdeplot(ori_data[:, :, feature_idx].flatten(), label='Original Data', shade=True)
    sns.kdeplot(generated_data[:, :, feature_idx].flatten(), label='Synthetic Data', shade=True)
    plt.title(f"Feature {feature_idx} Distribution")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

# 3. Box plots of features
for feature_idx in range(dim):
    data_to_plot = [ori_data[:, :, feature_idx].flatten(), generated_data[:, :, feature_idx].flatten()]
    plt.figure(figsize=(6, 4))
    plt.boxplot(data_to_plot, labels=['Original', 'Synthetic'])
    plt.title(f"Feature {feature_idx} Box Plot")
    plt.ylabel("Value")
    plt.show()

# 4. Heatmaps of correlation matrices
def plot_correlation_heatmap(data, title):
    # data shape: [no, seq_len, dim]
    # Reshape to [no * seq_len, dim]
    reshaped_data = data.reshape(-1, dim)
    corr = np.corrcoef(reshaped_data, rowvar=False)
    plt.figure(figsize=(8, 6))
    sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
    plt.title(title)
    plt.show()

plot_correlation_heatmap(ori_data, "Original Data Feature Correlation")
plot_correlation_heatmap(generated_data, "Synthetic Data Feature Correlation")

# 5. PCA Visualization
from sklearn.decomposition import PCA

def plot_pca(data1, data2, title1='Original Data', title2='Synthetic Data'):
    # Combine data
    data1_reshaped = data1.reshape(-1, dim)
    data2_reshaped = data2.reshape(-1, dim)
    pca = PCA(n_components=2)
    pca.fit(np.concatenate([data1_reshaped, data2_reshaped], axis=0))
    data1_pca = pca.transform(data1_reshaped)
    data2_pca = pca.transform(data2_reshaped)

    plt.figure(figsize=(8, 6))
    plt.scatter(data1_pca[:, 0], data1_pca[:, 1], label=title1, alpha=0.5)
    plt.scatter(data2_pca[:, 0], data2_pca[:, 1], label=title2, alpha=0.5)
    plt.title("PCA Visualization")
    plt.xlabel("Principal Component 1")
    plt.ylabel("Principal Component 2")
    plt.legend()
    plt.show()

plot_pca(ori_data, generated_data)

# 6. Time series line plots for individual features across all sequences
for feature_idx in range(dim):
    plt.figure(figsize=(12, 6))
    plt.plot(np.mean(ori_data[:, :, feature_idx], axis=0), label='Original Data Mean', color='blue')
    plt.plot(np.mean(generated_data[:, :, feature_idx], axis=0), label='Synthetic Data Mean', color='red', linestyle='dashed')
    plt.fill_between(range(seq_len),
                     np.mean(ori_data[:, :, feature_idx], axis=0) - np.std(ori_data[:, :, feature_idx], axis=0),
                     np.mean(ori_data[:, :, feature_idx], axis=0) + np.std(ori_data[:, :, feature_idx], axis=0),
                     color='blue', alpha=0.2)
    plt.fill_between(range(seq_len),
                     np.mean(generated_data[:, :, feature_idx], axis=0) - np.std(generated_data[:, :, feature_idx], axis=0),
                     np.mean(generated_data[:, :, feature_idx], axis=0) + np.std(generated_data[:, :, feature_idx], axis=0),
                     color='red', alpha=0.2)
    plt.title(f"Mean and Std Dev of Feature {feature_idx} Over Time")
    plt.xlabel("Time Step")
    plt.ylabel("Value")
    plt.legend()
    plt.show()

# 7. Statistical comparison
print("Original Data Mean per Feature:")
print(np.mean(ori_data, axis=(0, 1)))
print("Synthetic Data Mean per Feature:")
print(np.mean(generated_data, axis=(0, 1)))

print("Original Data Std Dev per Feature:")
print(np.std(ori_data, axis=(0, 1)))
print("Synthetic Data Std Dev per Feature:")
print(np.std(generated_data, axis=(0, 1)))

# 8. Compute and display the Mean Squared Error and Mean Absolute Error per feature
from sklearn.metrics import mean_squared_error, mean_absolute_error

for feature_idx in range(dim):
    mse = mean_squared_error(ori_data[:, :, feature_idx].flatten(), generated_data[:, :, feature_idx].flatten())
    mae = mean_absolute_error(ori_data[:, :, feature_idx].flatten(), generated_data[:, :, feature_idx].flatten())
    print(f"Feature {feature_idx}: MSE = {mse}, MAE = {mae}")
```

```python
#created better algorithms of distance calculation

# -*- coding: utf-8 -*-
"""
Created on Fri Jan 15 15:52:59 2021

Refactored by [Your Name]

Updated on [Today's Date]
"""

# -------------------------------
# Importing Required Libraries
# -------------------------------

import torch
import numpy as np
import os
from scipy.stats import wasserstein_distance
from sklearn.metrics import precision_recall_curve, auc
from sklearn.neighbors import KernelDensity, NearestNeighbors
from sklearn.decomposition import PCA
from scipy.linalg import sqrtm
import matplotlib.pyplot as plt
import seaborn as sns

# Enable inline plotting for Jupyter notebooks (e.g., Google Colab)
%matplotlib inline

# -------------------------------
# Device Configuration
# -------------------------------

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

# -------------------------------
# Define Metric Functions
# -------------------------------

def feature_distribution(X, Y):
    """
    Compute basic statistical features (mean and standard deviation) for datasets X and Y.

    Parameters:
    - X (np.ndarray): First dataset.
    - Y (np.ndarray): Second dataset.

    Returns:
    - dict: Dictionary containing means and standard deviations.
    """
    feat_dist = {
        'X_mean': np.mean(X, axis=0),
        'X_std': np.std(X, axis=0),
        'Y_mean': np.mean(Y, axis=0),
        'Y_std': np.std(Y, axis=0)
    }
    return feat_dist

def compute_wd(X, Y, params=None):
    """
    Compute Wasserstein Distance between two datasets.

    Parameters:
    - X (np.ndarray): First dataset.
    - Y (np.ndarray): Second dataset.
    - params (dict, optional): Parameters for Wasserstein Distance computation.

    Returns:
    - float: Wasserstein Distance.
    """
    # Flatten the data if it's multidimensional
    X_flat = X.flatten()
    Y_flat = Y.flatten()
    wd = wasserstein_distance(X_flat, Y_flat)
    return wd

def compute_frechet_distance(X, Y):
    """
    Compute Frechet Inception Distance (FID) between two datasets.

    Parameters:
    - X (np.ndarray): First dataset (e.g., real data).
    - Y (np.ndarray): Second dataset (e.g., generated data).

    Returns:
    - float: Frechet Distance.
    """
    # Compute the mean and covariance of each dataset
    mu1 = np.mean(X, axis=0)
    sigma1 = np.cov(X, rowvar=False)
    mu2 = np.mean(Y, axis=0)
    sigma2 = np.cov(Y, rowvar=False)

    # Compute the difference between means
    diff = mu1 - mu2

    # Compute the square root of the product of covariances
    covmean, _ = sqrtm(sigma1.dot(sigma2), disp=False)
    if np.iscomplexobj(covmean):
        covmean = covmean.real

    # Compute the Frechet Distance
    fid = diff.dot(diff) + np.trace(sigma1 + sigma2 - 2 * covmean)
    return fid

def compute_parzen(X, Y, sigma=0.408):
    """
    Compute Parzen Log-Likelihood of Y given X using Kernel Density Estimation.

    Parameters:
    - X (np.ndarray): Training dataset.
    - Y (np.ndarray): Test dataset.
    - sigma (float): Bandwidth for the Gaussian kernel.

    Returns:
    - tuple: (log-likelihood, standard error)
    """
    kde = KernelDensity(kernel='gaussian', bandwidth=sigma)
    kde.fit(X)
    log_likelihood = kde.score_samples(Y)
    parzen_ll = np.mean(log_likelihood)
    parzen_std = np.std(log_likelihood) / np.sqrt(len(Y))
    return parzen_ll, parzen_std

def compute_prc(X, Y):
    """
    Compute Precision-Recall Curve between two datasets.

    Parameters:
    - X (np.ndarray): Ground truth dataset.
    - Y (np.ndarray): Predicted dataset.

    Returns:
    - dict: Dictionary containing precision, recall, and average precision.
    """
    # Labels: X as 1 (real), Y as 0 (generated)
    y_true = np.concatenate([np.ones(len(X)), np.zeros(len(Y))])
    y_scores = np.concatenate([np.ones(len(X)), np.zeros(len(Y))])  # Simplistic scores

    precision, recall, _ = precision_recall_curve(y_true, y_scores)
    avg_precision = auc(recall, precision)
    prc = {
        'precision': precision,
        'recall': recall,
        'average_precision': avg_precision
    }
    return prc

def compute_prdc(X, Y):
    """
    Compute Precision, Recall, Density, and Coverage (PRDC) metrics.

    Parameters:
    - X (np.ndarray): Real dataset.
    - Y (np.ndarray): Generated dataset.

    Returns:
    - dict: Dictionary containing PRDC metrics.
    """
    # Precision and Recall can be approximated using nearest neighbors

    k = 5
    # Precision: How many generated samples are close to real samples
    nbrs_real = NearestNeighbors(n_neighbors=k, algorithm='auto').fit(X)
    distances, _ = nbrs_real.kneighbors(Y)
    precision = np.mean(distances.min(axis=1) < 1.0)  # Threshold can be adjusted

    # Recall: How many real samples are covered by generated samples
    nbrs_gen = NearestNeighbors(n_neighbors=k, algorithm='auto').fit(Y)
    distances, _ = nbrs_gen.kneighbors(X)
    recall = np.mean(distances.min(axis=1) < 1.0)  # Threshold can be adjusted

    # Density: Average distance of generated samples to real samples
    density = np.mean(distances.min(axis=1))

    # Coverage: Fraction of real samples covered by generated samples
    coverage = recall

    prdc = {
        'precision': precision,
        'recall': recall,
        'density': density,
        'coverage': coverage
    }
    return prdc

def compute_alpha_precision(X, Y, emb_center):
    """
    Compute One-Class Metrics: Alpha Precision and Beta Coverage.

    Parameters:
    - X (np.ndarray): In-class dataset.
    - Y (np.ndarray): Out-of-class dataset.
    - emb_center (np.ndarray): Center point for One-Class evaluation.

    Returns:
    - tuple: (alphas, alpha_precision_curve, beta_coverage_curve, Delta_precision_alpha, Delta_coverage_beta, authen)
    """
    # Placeholder implementation: You need to define what these metrics represent
    # For demonstration, we'll compute distance-based metrics

    # Compute distances from emb_center
    distances_X = np.linalg.norm(X - emb_center, axis=1)
    distances_Y = np.linalg.norm(Y - emb_center, axis=1)

    # Define alphas as quantiles
    alphas = np.linspace(0, 1, 100)
    alpha_precision_curve = []
    beta_coverage_curve = []
    Delta_precision_alpha = 0
    Delta_coverage_beta = 0
    authen = []

    for alpha in alphas:
        threshold = np.quantile(distances_X, alpha)
        precision = np.mean(distances_Y > threshold)
        recall = np.mean(distances_X <= threshold)
        alpha_precision_curve.append(precision)
        beta_coverage_curve.append(recall)
        Delta_precision_alpha += precision
        Delta_coverage_beta += recall
        authen.append(precision * recall)  # Example authenticity measure

    Delta_precision_alpha /= len(alphas)
    Delta_coverage_beta /= len(alphas)
    Delta_authenticity = np.mean(authen)

    return alphas, alpha_precision_curve, beta_coverage_curve, Delta_precision_alpha, Delta_coverage_beta, authen

def compute_identifiability(X, Y):
    """
    Compute Identifiability between two datasets.

    Parameters:
    - X (np.ndarray): First dataset.
    - Y (np.ndarray): Second dataset.

    Returns:
    - float: Identifiability score.
    """
    # Placeholder implementation: You need to define the actual metric
    # For demonstration, we'll compute the cosine similarity between means
    from sklearn.metrics.pairwise import cosine_similarity

    mu_X = np.mean(X, axis=0).reshape(1, -1)
    mu_Y = np.mean(Y, axis=0).reshape(1, -1)
    similarity = cosine_similarity(mu_X, mu_Y)[0][0]
    return similarity

# -------------------------------
# Define Visualization Functions
# -------------------------------

def plot_feature_distribution(feat_dist):
    """
    Plot histograms of feature means and standard deviations for X and Y.

    Parameters:
    - feat_dist (dict): Dictionary containing feature distribution statistics.
    """
    plt.figure(figsize=(12, 5))

    # Plot means
    plt.subplot(1, 2, 1)
    plt.hist(feat_dist['X_mean'], bins=30, alpha=0.5, label='X Mean')
    plt.hist(feat_dist['Y_mean'], bins=30, alpha=0.5, label='Y Mean')
    plt.title('Feature Means Distribution')
    plt.xlabel('Mean Value')
    plt.ylabel('Frequency')
    plt.legend()

    # Plot standard deviations
    plt.subplot(1, 2, 2)
    plt.hist(feat_dist['X_std'], bins=30, alpha=0.5, label='X Std')
    plt.hist(feat_dist['Y_std'], bins=30, alpha=0.5, label='Y Std')
    plt.title('Feature Standard Deviation Distribution')
    plt.xlabel('Standard Deviation')
    plt.ylabel('Frequency')
    plt.legend()

    plt.tight_layout()
    plt.show()

def plot_wasserstein_distance(X, Y):
    """
    Plot histograms of the two datasets to visualize their distributions.

    Parameters:
    - X (np.ndarray): First dataset.
    - Y (np.ndarray): Second dataset.
    """
    plt.figure(figsize=(10, 5))
    plt.hist(X.flatten(), bins=100, alpha=0.5, label='X', density=True)
    plt.hist(Y.flatten(), bins=100, alpha=0.5, label='Y', density=True)
    plt.title('Wasserstein Distance Visualization')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

def plot_frechet_distance(X, Y):
    """
    Plot scatter plots of the first two principal components of X and Y.

    Parameters:
    - X (np.ndarray): First dataset.
    - Y (np.ndarray): Second dataset.
    """
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X)
    Y_pca = pca.transform(Y)

    plt.figure(figsize=(8, 6))
    plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.5, label='X (Real)')
    plt.scatter(Y_pca[:, 0], Y_pca[:, 1], alpha=0.5, label='Y (Generated)')
    plt.title('Frechet Distance Visualization (PCA)')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.legend()
    plt.show()

def plot_parzen_log_likelihood(parzen_ll, parzen_std):
    """
    Plot the Parzen Log-Likelihood with error bars.

    Parameters:
    - parzen_ll (float): Log-likelihood.
    - parzen_std (float): Standard error of log-likelihood.
    """
    plt.figure(figsize=(6, 4))
    plt.bar(['Parzen Log-Likelihood'], [parzen_ll], yerr=[parzen_std], capsize=5, color='skyblue')
    plt.title('Parzen Log-Likelihood')
    plt.ylabel('Log-Likelihood')
    plt.show()

def plot_precision_recall(prc):
    """
    Plot the Precision-Recall Curve.

    Parameters:
    - prc (dict): Dictionary containing precision, recall, and average precision.
    """
    plt.figure(figsize=(6, 6))
    plt.plot(prc['recall'], prc['precision'], label=f'AP = {prc["average_precision"]:.2f}')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_prdc(prdc):
    """
    Plot a bar chart of Precision, Recall, Density, and Coverage.

    Parameters:
    - prdc (dict): Dictionary containing PRDC metrics.
    """
    metrics = ['precision', 'recall', 'density', 'coverage']
    values = [prdc[m] for m in metrics]

    plt.figure(figsize=(8, 6))
    sns.barplot(x=metrics, y=values, palette='viridis')
    plt.title('PRDC Metrics')
    plt.ylabel('Value')
    plt.ylim(0, 1.1)
    plt.show()

def plot_one_class_metrics(alphas, alpha_precision_curve, beta_coverage_curve):
    """
    Plot Alpha Precision and Beta Coverage curves.

    Parameters:
    - alphas (np.ndarray): Array of alpha values.
    - alpha_precision_curve (list): Precision values corresponding to alphas.
    - beta_coverage_curve (list): Coverage (Recall) values corresponding to alphas.
    """
    plt.figure(figsize=(10, 5))

    # Plot Alpha Precision Curve
    plt.subplot(1, 2, 1)
    plt.plot(alphas, alpha_precision_curve, label='Alpha Precision')
    plt.title('Alpha Precision Curve')
    plt.xlabel('Alpha')
    plt.ylabel('Precision')
    plt.legend()
    plt.grid(True)

    # Plot Beta Coverage Curve
    plt.subplot(1, 2, 2)
    plt.plot(alphas, beta_coverage_curve, label='Beta Coverage', color='orange')
    plt.title('Beta Coverage Curve')
    plt.xlabel('Alpha')
    plt.ylabel('Coverage (Recall)')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

def plot_identifiability(similarity):
    """
    Print and explain the Identifiability score.

    Parameters:
    - similarity (float): Cosine similarity score.
    """
    print(f"Identifiability (Cosine Similarity between means): {similarity:.4f}")
    if similarity > 0.8:
        print("High identifiability: The datasets have similar mean directions.")
    elif similarity > 0.5:
        print("Moderate identifiability: The datasets have somewhat similar mean directions.")
    else:
        print("Low identifiability: The datasets have dissimilar mean directions.")

# -------------------------------
# Define the compute_metrics Function
# -------------------------------

def compute_metrics(X, Y, which_metric=None, wd_params=None, model=None):
    """
    Compute various metrics between two datasets X and Y.

    Parameters:
    - X (np.ndarray): First dataset.
    - Y (np.ndarray): Second dataset.
    - which_metric (list of lists, optional): Metrics to compute for each embedding type.
    - wd_params (dict, optional): Parameters for Wasserstein Distance computation.
    - model (torch.nn.Module, optional): Model for embedding data (e.g., OneClass model).

    Returns:
    - results (dict): Computed metrics.
    """
    results = {}
    emb_types = ['']  # Default embedding type

    # Append '_OC' if a model is provided for OneClass embedding
    if model is not None:
        emb_types.append('_OC')
    else:
        print('#####################!OC model not defined !##################')

    # Set default Wasserstein Distance parameters if not provided
    if wd_params is None:
        wd_params = {
            'iterations': 500,
            'h_dim': 30,
            'z_dim': 10,
            'mb_size': 128
        }

    # Define default metrics if not specified
    if which_metric is None:
        which_metric = [
            ['WD', 'FD', 'PRDC', 'OC'],  # Metrics for default embedding
            ['OC']                        # Metrics for OneClass embedding
        ]

    # Iterate over each embedding type
    for emb_index, emb in enumerate(emb_types):
        print(f"\n--- Processing Embedding Type: '{emb}' ---")

        # Handle OneClass embedding if applicable
        if emb_index == 1 and len(which_metric[1]) > 0:
            print('Computing metrics for OC embedding')
            print('Embedding data into OC representation')
            model.to(device)
            model.eval()  # Set model to evaluation mode
            with torch.no_grad():
                X_tensor = torch.tensor(X).float().to(device)
                Y_tensor = torch.tensor(Y).float().to(device)
                X_emb = model(X_tensor).cpu().numpy()
                Y_emb = model(Y_tensor).cpu().numpy()
            print('Done embedding')
            print(f'X_emb: mean={np.mean(X_emb):.4f}, std={np.std(X_emb):.4f}')
            print(f'Y_emb: mean={np.mean(Y_emb):.4f}, std={np.std(Y_emb):.4f}')
            X_processed, Y_processed = X_emb, Y_emb
        else:
            print('Computing metrics for default embedding')
            X_processed, Y_processed = X, Y

        # -------------------------------
        # 1. Marginal Feature Distributions
        # -------------------------------
        if 'marg' in which_metric[emb_index]:
            print('Start computing marginal feature distributions')
            try:
                feat_dist = feature_distribution(X_processed, Y_processed)
                results[f'feat_dist{emb}'] = feat_dist
                print('Finished computing feature distributions')
                print(feat_dist)

                # Plot feature distributions
                plot_feature_distribution(feat_dist)
            except Exception as e:
                print(f"Error computing feature distributions: {e}")

        # -------------------------------
        # 2. Wasserstein Distance (WD)
        # -------------------------------
        if 'WD' in which_metric[emb_index]:
            print('Start computing Wasserstein Distance')
            try:
                wd_measure = compute_wd(X_processed, Y_processed, wd_params)
                results[f'wd_measure{emb}'] = wd_measure
                print(f'Wasserstein Distance: {wd_measure}')

                # Plot Wasserstein Distance visualization
                plot_wasserstein_distance(X_processed, Y_processed)
            except Exception as e:
                print(f"Error computing Wasserstein Distance: {e}")

        # -------------------------------
        # 3. Identifiability
        # -------------------------------
        if 'ID' in which_metric[emb_index]:
            print('Start computing Identifiability')
            try:
                identifiability = compute_identifiability(X_processed, Y_processed)
                results[f'identifiability{emb}'] = identifiability
                print(f'Identifiability Measure: {identifiability}')

                # Plot Identifiability
                plot_identifiability(identifiability)
            except Exception as e:
                print(f"Error computing Identifiability: {e}")

        # -------------------------------
        # 4. Frechet Distance
        # -------------------------------
        if 'FD' in which_metric[emb_index] or 'FID' in which_metric[emb_index]:
            print('Start computing Frechet Distance')
            try:
                fid_value = compute_frechet_distance(X_processed, Y_processed)
                results[f'fid_value{emb}'] = fid_value
                fid_per_dim = fid_value / Y_processed.shape[-1]
                print(f'Frechet Distance: {fid_value}')
                print(f'Frechet Distance per Dimension: {fid_per_dim}')

                # Plot Frechet Distance visualization
                plot_frechet_distance(X_processed, Y_processed)
            except Exception as e:
                print(f"Error computing Frechet Distance: {e}")

        # -------------------------------
        # 5. Parzen Log-Likelihood
        # -------------------------------
        if 'parzen' in which_metric[emb_index]:
            print('Start computing Parzen Log-Likelihood')
            try:
                parzen_ll, parzen_std = compute_parzen(X_processed, Y_processed, sigma=0.408)
                results[f'parzen_ll{emb}'] = parzen_ll
                results[f'parzen_std{emb}'] = parzen_std
                print(f'Parzen Log-Likelihood of test set = {parzen_ll}, SE = {parzen_std}')

                # Plot Parzen Log-Likelihood
                plot_parzen_log_likelihood(parzen_ll, parzen_std)
            except Exception as e:
                print(f"Error computing Parzen Log-Likelihood: {e}")

        # -------------------------------
        # 6. Precision/Recall Metrics
        # -------------------------------
        if 'PR' in which_metric[emb_index]:
            print('Start computing Precision/Recall')
            try:
                prc = compute_prc(X_processed, Y_processed)
                results[f'PR{emb}'] = prc
                print(f'Precision/Recall: {prc}')

                # Plot Precision-Recall Curve
                plot_precision_recall(prc)
            except Exception as e:
                print(f"Error computing Precision/Recall: {e}")
        elif 'PRDC' in which_metric[emb_index]:
            print('Start computing Precision, Recall, Density, and Coverage (PRDC)')
            try:
                prdc_res = compute_prdc(X_processed, Y_processed)
                for key, value in prdc_res.items():
                    results[f'{key}{emb}'] = value
                    print(f'PRDC - {key}: {value}')

                # Plot PRDC Metrics
                plot_prdc(prdc_res)
            except Exception as e:
                print(f"Error computing PRDC: {e}")

        # -------------------------------
        # 7. OneClass Metrics
        # -------------------------------
        if 'OC' in which_metric[emb_index]:
            print('Start computing OneClass Metrics')
            try:
                if emb_index == 1:
                    # Assuming 'model.c' is the center for OC embedding
                    emb_center = model.c.cpu().numpy()
                else:
                    emb_center = np.mean(X_processed, axis=0)

                OC_res = compute_alpha_precision(X_processed, Y_processed, emb_center)
                alphas, alpha_precision_curve, beta_coverage_curve, Delta_precision_alpha, Delta_coverage_beta, authen = OC_res

                results[f'alphas{emb}'] = alphas
                results[f'alpha_pc{emb}'] = alpha_precision_curve
                results[f'beta_cv{emb}'] = beta_coverage_curve
                results[f'auten{emb}'] = authen
                results[f'Dpa{emb}'] = Delta_precision_alpha
                results[f'Dcb{emb}'] = Delta_coverage_beta
                results[f'Daut{emb}'] = np.mean(authen)

                print(f'OneClass - Delta_precision_alpha: {results[f"Dpa{emb}"]}')
                print(f'OneClass - Delta_coverage_beta: {results[f"Dcb{emb}"]}')
                print(f'OneClass - Delta_authenticity: {results[f"Daut{emb}"]}')

                # Plot One-Class Metrics
                plot_one_class_metrics(alphas, alpha_precision_curve, beta_coverage_curve)
            except AttributeError as ae:
                print(f"Attribute Error in OneClass Metrics: {ae}")
            except Exception as e:
                print(f"Error computing OneClass Metrics: {e}")

    return results

# -------------------------------
# Example Usage (Main Section)
# -------------------------------

if __name__ == "__main__":
    # Example data (replace with actual data)
    X = np.random.randn(1000, 64)  # Example feature vectors for dataset X
    Y = np.random.randn(1000, 64)  # Example feature vectors for dataset Y

    # Example model (replace with your actual model)
    class ExampleModel(torch.nn.Module):
        def __init__(self, input_dim=64, output_dim=32):
            super(ExampleModel, self).__init__()
            self.fc = torch.nn.Linear(input_dim, output_dim)
            self.c = torch.randn(output_dim)  # Example center for OneClass

        def forward(self, x):
            return self.fc(x)

    model = ExampleModel()

    # Define which metrics to compute (optional)
    which_metric = [
        ['WD', 'FD', 'PRDC', 'OC'],  # Metrics for default embedding
        ['OC']                        # Metrics for OneClass embedding
    ]

    # Define Wasserstein Distance parameters (optional)
    wd_params = {
        'iterations': 500,
        'h_dim': 30,
        'z_dim': 10,
        'mb_size': 128
    }

    # Compute metrics
    metrics_results = compute_metrics(X, Y, which_metric=which_metric, wd_params=wd_params, model=model)

    # Display results
    print("\n=== Computed Metrics ===")
    for metric, value in metrics_results.items():
        print(f"{metric}: {value}")
```
